```python
df1=pd.read_excel('X.xlsx')
df2=pd.read_excel('y.xlsx')
```

```python
df=pd.concat([df1,df2],axis=1)
df
```

```python
df['Expenditure']=df['Safety Features']+df['Tech Features']+df['Marketing Spend']
df
```

```python
X=df.iloc[:,-1].values
y=df.iloc[:,-2].values
X,y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
s_model = LinearRegression().fit(X_train.reshape(-1,1), y_train)
```

```python
print('coefficient of determination:', s_model.score(X_train.reshape(-1,1), y_train))
print('intercept:', s_model.intercept_)
print('slope:', s_model.coef_)
```

```python
y_pred=s_model.predict(X_test.reshape(-1,1))
X_test,y_test,y_pred
```

```python
y_pred2=s_model.predict(X_train.reshape(-1,1))
X_train,y_train,y_pred2
```

```python
r2_score(y_test,y_pred)
```
