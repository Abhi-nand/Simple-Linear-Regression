```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('Mtcars_Data_Practice.csv')
df
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('xlabel', fontsize=12)
ax.set_ylabel('ylabel', fontsize=12)
df.plot.scatter(x = 'disp', y = 'hp', s = 20, ax=ax);
plt.show()
```

```python
X=np.array(df['disp']).reshape((-1, 1))
y=np.array(df['hp'])
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=200)
```

```python
s_model = LinearRegression().fit(X_train, y_train)
s_r_sq=s_model.score(X_train, y_train)
```

```python
print('coefficient of determination:', s_r_sq)
print('intercept:', s_model.intercept_)
print('slope:', s_model.coef_)
```

```python
y_pred = s_model.predict(X_test)
print('predicted response:', y_pred, sep='\n')
```

```python
MSE=mean_squared_error(y_test, y_pred)
MAE=mean_absolute_error(y_test,y_pred)
RMSE = mean_squared_error(y_test, y_pred, squared=False)
print(MSE, MAE, RMSE)
```
